# Mandelbrot Set in Python

The Mandelbrot set is quite interesting mathematically. Let's first explore what the set is. For a complex number $ c = a + bi $ in the complex plane, the Mandelbrot set is defined as all of the values of $ c $ for which the complex function $$ f_c (z) = z^{2} + c $$ remains absolutely bounded under iteration for $ z = 0 $ (i.e. $f_c(0), f_c(f_c(0))$ ...) remains bounded. Plotted on the complex plane, the Mandelbrot set looks quite elegant and has some interesting properties of self-similarity. Let's see if we can generate an elementary plot of the set.

```python
from pylab import *
from numpy import NaN
 
#define the function, n = 100 iterations
def mandelbrot(c):
	z = 0
	for i in range(1, 100):
		z = z**2 + c
		if abs(z) > 2:
			return i
	return NaN
 
X = arange(-2, .5, .002)
Y = arange(-1,  1, .002)
Z = zeros((len(Y), len(X)))

#generate datapoints
for iy, y in enumerate(Y):
	for ix, x in enumerate(X):
		Z[iy,ix] = m(x + 1j * y)

#plot 

imshow(Z, cmap = plt.cm.winter, interpolation = 'none', extent = (X.min(), X.max(), Y.min(), Y.max()))
xlabel("Re(c)")
ylabel("Im(c)")

show()
```

There we have it. A basic plot o the Mandelbrot set. Pretty cool!
